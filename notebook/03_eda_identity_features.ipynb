```python
import sys
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
plt.style.use("seaborn-v0_8-darkgrid")

DATA_DIR = Path("../data/interim")
TRAIN_PATH = DATA_DIR / "train_merged.parquet"
```

## EDA - Identity Features

## Context

The identity table provides device and browser fingerprinting data for a subset of transactions. Understanding this auxiliary data source helps us design features that leverage identity signals when available, while also handling cases where this data is missing.

## Objectives

- Quantify identity table coverage and examine its relationship to fraud patterns
- Analyze device-related features (DeviceType, DeviceInfo) and their distributions
- Explore id_* features and identify their associations with fraudulent transactions
- Develop robust strategies for handling identity-related missing data

## Approach

We'll investigate how the presence or absence of identity information correlates with fraud rates, extract meaningful patterns from device fingerprints, and create features that work effectively even when identity data is incomplete.

```python
train = pd.read_parquet(TRAIN_PATH)
print(f'Data loaded: {train.shape}')
```

```python
print(f'Total columns: {len(train.columns)}')
print(f'Identity columns : {len([x for x in train.columns if x.startswith("id_")])}')
```

```python
# kolom identitas
id_cols = [c for c in train.columns if c.startswith('id_')]
device_cols = ['DeviceType', 'DeviceInfo']

# cek apakah ada data identitas di tiap baris
identity = train[id_cols + device_cols].notna().any(axis=1)

print('Identity Coverage')
print(f'Total Transaction : {len(train):,}')
print(f'   With identity: {identity.sum():,} ({identity.mean()*100:.1f}%)')
print(f'   Without identity: {(~identity).sum():,} ({(~identity).mean()*100:.1f}%)')
```

### Identity Coverage


```python
# fraud rate berdasarkan kehadiran data identitas

fraud_id = train.loc[identity, 'isFraud'].mean() * 100
fraud_without_id = train.loc[~identity, 'isFraud'].mean() * 100

print('Fraud Rate by identity presence')
print(f'with identity : {fraud_id:.2f}%')
print(f'without identity : {fraud_without_id:.2f}%')

# visualisasi perbedaan fraud rate
fig, ax = plt.subplots(figsize=(8,6))
ax.bar(['With Identity', 'Without Identity'], [fraud_id, fraud_without_id])
ax.set_ylabel('Fraud Rate (%) ')
ax.set_title('Fraud by Identity Presence')
ax.axhline(y=train['isFraud'].mean() * 100, linestyle='--', label='Overall Rate')
ax.legend()
plt.tight_layout()
plt.show()
```

## DEVICE TYPE


```python
print('DeviceType Distribution:')
print(train['DeviceType'].value_counts(dropna=False))

device_fraud = train.groupby('DeviceType', dropna=False)['isFraud'].agg(['mean', 'count'])
device_fraud['mean'] *= 100

print('Fraud Rate by DeviceType:')
print(device_fraud)
```

```python
top_device = train['DeviceInfo'].value_counts().head(5)

fig, ax = plt.subplots(figsize = (10 , 5))
top_device.plot(kind='bar', color='orange')
ax.set_ylabel('count', fontsize=15)
plt.xticks(rotation=45, ha='right')
```

```python
# eksplorasi DeviceInfo
# top 15 device info yang paling sering muncul
top_device = train['DeviceInfo'].value_counts().head(15)

print('top 15 device info')
print(top_device)

# fraud rate untuk masing-masing device info 
fraud_by_device = train.groupby('DeviceInfo')['isFraud'].mean() * 100
fraud_by_device = fraud_by_device.loc[top_device.index]

fig, ax = plt.subplots(figsize=(10,6))
fraud_by_device.sort_values(ascending=False).plot(kind='bar', ax=ax)
ax.set_title('Fraud Rate for Top Device Info')
ax.set_ylabel('Fraud Rate (%)')
ax.axhline(y=train['isFraud'].mean() * 100, color='red', linestyle='--', label='Overal Rate')
ax.legend()
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## ID FEATURES

```python
id_cols = [f'id_{i:02d}' for i in range(1,39)]
id_cols = [c for c in id_cols if c in train.columns]

print(f'Id Features {len(id_cols)} columns')

# hitung missing rate dan jumlah nilai unik tiap fitur ID 
id_stats = []
for col in id_cols:
    id_stats.append({
        'columns' : col,
        'missing_pct' : train[col].isnull().sum() / len(train) * 100,
        'nunique' : train[col].nunique(),
        'dtype' : str(train[col].dtype)
    })

id_df = pd.DataFrame(id_stats)
print(id_df.to_string())
```

```python
# korelasi fitur ID numerik terhadap target isFraud

numeric_id_cols = train[id_cols].select_dtypes(include=[np.number]).columns.tolist()

if numeric_id_cols:
    id_corr = train[numeric_id_cols + ['isFraud']].corr()['isFraud'].drop('isFraud').sort_values()

    print('id features correlations with isfraud :')
    print(id_corr)

    fig, ax = plt.subplots(figsize=(10,6))
    id_corr.plot(kind='barh', ax = ax)
    ax.set_title('ID features correlation with isFraud')
    plt.xlabel('correlations')
    ax.axvline(x=0, color='black', linestyle='-')
    plt.tight_layout()
    plt.show()
    
```

## Key Takeaways

1. **Identity coverage selective (~25%)**: Bikin `has_identity` sebagai core feature
2. **Device type matters**: Mobile vs Desktop menunjukkan pola fraud yang berbeda
3. **DeviceInfo perlu cleaning**: Extract browser family, OS components sebelum dipake
4. **ID features valuable tapi opaque**: Pake dengan proper encoding techniques

## Summary

Ringkasan identity features EDA:

- **Identity coverage**: ~25% transaksi punya identity data. Identity presence itself predictive
- **DeviceType**: Mobile vs desktop beda fraud rate
- **DeviceInfo**: High cardinality — extract browser family, jangan raw values
- **ID features**: 38 features dengan ~75% missingness
- **Feature to engineer**: `has_identity` binary flag, `browser_family`, `is_mobile`
